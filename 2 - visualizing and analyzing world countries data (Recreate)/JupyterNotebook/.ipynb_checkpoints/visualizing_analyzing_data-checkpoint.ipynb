```python
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
df
```

```python
df['continent'].unique()
```

```python
df1 = df.groupby(['country', 'continent'])[['pop', 'lifeExp', 'gdpPercap']].sum().reset_index()
df1
```

```python
df2 = df1[df1['continent'] == 'Asia'].nlargest(10, columns=['lifeExp'])
df2
```

```python
top_continent = df2['continent'].iloc[0]
top_continent
```
